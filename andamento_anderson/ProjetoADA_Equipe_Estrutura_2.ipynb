## - monthly_revenue.csv – Base de dados contendo o total de vendas mensal, em R$. 

```python
# Total de vendes por semana
sum_weekly = df_all_sales_refined.groupby('n_week').agg('sum').iloc[:,2:]
sum_weekly
```

```python
# Copiando Base
price_weekly = pd.read_csv('weekly_price.csv', sep=",", index_col='n_week')
price_weekly
```

```python
# Total por semana = Preco x Volume
df_monthly_revenue_week = (price_weekly * sum_weekly).round(2)
df_monthly_revenue_week['n_week'] = df_monthly_revenue_week.index
df_monthly_revenue_week.reset_index(drop=True, inplace=True)
df_monthly_revenue_week
```
